# Patr贸n de Dise帽o Strategy

## Introducci贸n

El patr贸n de dise帽o Strategy es un patr贸n de comportamiento que permite definir una familia de algoritmos, encapsular cada uno de ellos y hacerlos intercambiables. Este patr贸n permite que el algoritmo var铆e independientemente de los clientes que lo utilizan.

## Explicaci贸n Detallada

### Definici贸n

- **Strategy**: Define una familia de algoritmos, encapsula cada uno de ellos y los hace intercambiables. Debido a que los algoritmos deben recibir los mismos par谩metros y devolver el mismo tipo de resultado, se pueden intercambiar sin que el cliente se d茅 cuenta.

### Beneficios del Patr贸n Strategy

1. **Flexibilidad**: Permite cambiar los algoritmos en tiempo de ejecuci贸n.

2. **Mantenibilidad**: Facilita la adici贸n de nuevos algoritmos sin modificar el c贸digo existente.

3. **Reusabilidad**: Promueve la reutilizaci贸n de algoritmos en diferentes contextos.

### Ejemplo Antes de Aplicar el Patr贸n Strategy

Supongamos que tenemos una clase `Route` que calcula la mejor ruta, el costo y el tiempo de viaje dependiendo del tipo de veh铆culo. Aqu铆 est谩 el c贸digo antes de aplicar el patr贸n Strategy:

```python
class Route:
    def __init__(self, origin: int, destination: int) -> None:
        self.origin: int = origin
        self.destination: int = destination

    def get_best_route(self, vehicle: str) -> dict:
        if vehicle == "car":
            return {"start_street": self.origin, "end_street": self.destination, "route": "Use the avenue 123"}
        elif vehicle == "bike":
            return {"start_street": self.origin, "end_street": self.destination, "route": "Use the bike lane"}
        return {}

    def get_cost(self, vehicle: str) -> float:
        if vehicle == "car":
            return round((self.destination - self.origin) * 0.1, 2)
        elif vehicle == "bike":
            return 0
        return 0

    def get_time(self, vehicle: str) -> float:
        if vehicle == "car":
            return round((self.destination - self.origin) * 0.5, 2)
        elif vehicle == "bike":
            return round((self.destination - self.origin) * 2, 2)
        return 0

    def avg_speed(self, vehicle: str) -> float:
        return round((self.destination - self.origin) / self.get_time(vehicle), 2)
```

#### An谩lisis del Ejemplo Incorrecto

- Tiene m煤ltiples responsabilidades y debe ser modificada cada vez que se agrega un nuevo tipo de veh铆culo.

- Viola el principio de `Single Responsibility`, ya que, tiene m谩s de una raz贸n para cambiar.

- Viola el principio de `Open/Closed`, ya que, no se puede extender sin modificar el c贸digo existente.

- Viola el principio de `Liskov Substitution`, ya que, no se puede reemplazar un tipo de veh铆culo por otro sin cambiar el comportamiento de la clase.

- Viola el principio de `Interface Segregation`, ya que, los m茅todos `calculate_best_route`, `calculate_cost` y `calculate_time` no son relevantes para todos los tipos de veh铆culos.

- Viola el principio de `Dependency Inversion`, ya que, depende de las implementaciones concretas de los tipos de veh铆culos.

- Hace que el c贸digo sea dif铆cil de mantener y extender.

### Ejemplo Despu茅s de Aplicar el Patr贸n Strategy

Aplicando el patr贸n Strategy, podemos refactorizar el c贸digo de la siguiente manera:

```python
from abc import ABC, abstractmethod

class RouteStrategy(ABC):
    @abstractmethod
    def get_best_route(self, origin: int, destination: int) -> dict:
        pass

    @abstractmethod
    def get_cost(self, origin: int, destination: int) -> float:
        pass

    @abstractmethod
    def get_time(self, origin: int, destination: int) -> float:
        pass
```

```python
class CarStrategy(RouteStrategy):
    def get_best_route(self, origin: int, destination: int) -> dict:
        return {"start_street": origin, "end_street": destination, "route": "Use the avenue 123"}

    def get_cost(self, origin: int, destination: int) -> float:
        return round((destination - origin) * 0.1, 2)

    def get_time(self, origin: int, destination: int) -> float:
        return round((destination - origin) * 0.5, 2)

class BikeStrategy(RouteStrategy):
    def get_best_route(self, origin: int, destination: int) -> dict:
        return {"start_street": origin, "end_street": destination, "route": "Use the bike lane"}

    def get_cost(self, origin: int, destination: int) -> float:
        return 0

    def get_time(self, origin: int, destination: int) -> float:
        return round((destination - origin) * 2, 2)
```

```python
class Route:
    def __init__(self, origin: int, destination: int, strategy: RouteStrategy) -> None:
        self.origin: int = origin
        self.destination: int = destination
        self.strategy: RouteStrategy = strategy

    def get_best_route(self) -> dict:
        return self.strategy.get_best_route(origin=self.origin, destination=self.destination)

    def get_cost(self) -> float:
        return self.strategy.get_cost(origin=self.origin, destination=self.destination)

    def get_time(self) -> float:
        return self.strategy.get_time(origin=self.origin, destination=self.destination)

    def avg_speed(self) -> float:
        return round((self.destination - self.origin) / self.get_time(), 2)
```

```python
# Ejemplo de uso
car_strategy = CarStrategy()
bike_strategy = BikeStrategy()

route_by_car = Route(origin=0, destination=10, strategy=car_strategy)
route_by_bike = Route(origin=0, destination=10, strategy=bike_strategy)

print("With car the route is:", route_by_car.get_best_route())
print("With car the cost is:", route_by_car.get_cost())
print("With car the time is:", route_by_car.get_time())
print("With car the average speed is:", route_by_car.avg_speed())

print("With bike the route is:", route_by_bike.get_best_route())
print("With bike the cost is:", route_by_bike.get_cost())
print("With bike the time is:", route_by_bike.get_time())
print("With bike the average speed is:", route_by_bike.avg_speed())
```

 Escenario: Sistema de c谩lculo de precio din谩mico para productos del hogar
Una compa帽铆a vende distintos productos para el hogar y quiere aplicar estrategias de precio seg煤n el tipo de cliente:

Cliente regular: paga el precio completo.

Cliente VIP: recibe un 20% de descuento.

Cliente institucional: paga con impuestos incluidos.


 Objetivo del ejercicio
Crear un sistema que permita cambiar la estrategia de c谩lculo de precio sin modificar la clase principal del producto. Cada estrategia ser谩 una clase independiente.

```python
#Estructura del ejercicio en Python


#1. Interfaz Strategy
from abc import ABC, abstractmethod

class EstrategiaPrecio(ABC):
    @abstractmethod
    def calcular_precio(self, producto, cantidad):
        pass


#2. Estrategias concretas
class PrecioRegular(EstrategiaPrecio):
    def calcular_precio(self, producto, cantidad):
        return producto.precio * cantidad

class PrecioVIP(EstrategiaPrecio):
    def calcular_precio(self, producto, cantidad):
        return producto.precio * cantidad * 0.8

class PrecioMayorista(EstrategiaPrecio):
    def calcular_precio(self, producto, cantidad):
        if cantidad >= 100:
            return producto.precio * cantidad * 0.7
        return producto.precio * cantidad

class PrecioInstitucional(EstrategiaPrecio):
    def calcular_precio(self, producto, cantidad):
        return producto.precio * cantidad * 1.19  # 19% impuesto
    

#3. Clase Producto con estrategia intercambiable
class Producto:
    def __init__(self, nombre, precio, estrategia: EstrategiaPrecio):
        self.nombre = nombre
        self.precio = precio
        self.estrategia = estrategia

    def obtener_precio_final(self, cantidad):
        return self.estrategia.calcular_precio(self, cantidad)


#4. Ejercicio de aplicaci贸n
aspiradora = Producto("Aspiradora CleanPro", 300000, PrecioVIP())
refrigerador = Producto("Refrigerador CoolTech", 800000, PrecioMayorista())
purificador = Producto("Purificador AirMax", 500000, PrecioInstitucional())

print("Precio final aspiradora (5 unidades):", aspiradora.obtener_precio_final(5))
print("Precio final refrigerador (120 unidades):", refrigerador.obtener_precio_final(120))
print("Precio final purificador (10 unidades):", purificador.obtener_precio_final(10))
```
